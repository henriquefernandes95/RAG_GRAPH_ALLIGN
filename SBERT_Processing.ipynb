```python
import pandas as pd
import rdflib
from collections import defaultdict

# generate_dataframe: uses simple SPARQL (rdflib-compatible) and groups results in Python
def generate_dataframe(datasetOnto: rdflib.graph.Graph, show_progress: bool = False, max_rows: int | None = None) -> pd.DataFrame:
    """
    Query structure expected by this function (rdflib SPARQL):
      PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
      PREFIX owl: <http://www.w3.org/2002/07/owl#>
      PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
      SELECT ?e ?p ?o ?p1 ?o1
      WHERE {
          ?e rdf:type owl:Class.
          ?e ?p ?o.
          ?o ?p1 ?o1.
      }
    """
    sparql = """
      PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
      PREFIX owl: <http://www.w3.org/2002/07/owl#>
      PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
      SELECT ?e ?p ?o ?p1 ?o1
      WHERE {
          ?e rdf:type owl:Class.
          ?e ?p ?o.
          ?o ?p1 ?o1.
      }
    """
    qres = datasetOnto.query(sparql)
    # optional progress wrapper (tqdm) - non-fatal if tqdm not installed
    iterator = qres
    if show_progress:
        try:
            import tqdm
            iterator = tqdm.tqdm(qres)
        except Exception:
            iterator = qres
    acc = defaultdict(lambda: {'entity': None, 'contextLevl0': [], 'contextLevl1': []})
    count = 0
    for row in iterator:
        # safe conversions (avoid None errors and expensive str() calls until needed)
        e = str(row.e) if row.e is not None else ''
        p = str(row.p) if getattr(row, 'p', None) is not None else ''
        o = str(row.o) if getattr(row, 'o', None) is not None else ''
        p1 = str(row.p1) if getattr(row, 'p1', None) is not None else ''
        o1 = str(row.o1) if getattr(row, 'o1', None) is not None else ''
        acc[e]['entity'] = e
        acc[e]['contextLevl0'].append(p + '\t' + o + '\n')
        acc[e]['contextLevl1'].append(o + '\t' + p1 + '\t' + o1)
        count += 1
        if max_rows is not None and count >= max_rows:
            break
    # build DataFrame once from accumulated data
    rows = []
    for e, vals in acc.items():
        rows.append({
            'entity': vals['entity'],
            'contextLevl0': ''.join(vals['contextLevl0']),
            'contextLevl1': 'FIM'.join(vals['contextLevl1'])
        })
    return pd.DataFrame.from_records(rows, columns=['entity', 'contextLevl0', 'contextLevl1'])

# generate both dataframes (no progress by default)
dFrameOrigin = generate_dataframe(datasetGraphNcit).reset_index(drop=True)
dFrameTarget = generate_dataframe(datasetGraphDoid).reset_index(drop=True)
```

```python
import pandas as pd # Ensure pandas is imported if this cell is run independently
import pickle
import re
from collections import defaultdict # Needed for the new mapping structure

try:
    results
except NameError:
    # Load from pickle file
    with open("items.pkl", "rb") as f:
        results = pickle.load(f)

# First, let's inspect the structure of results
print(f"Results type: {type(results)}")
print(f"Results length: {len(results)}")
print("\nFirst item structure:")
print(f"Type: {type(results[0])}")
print(f"Length: {len(results[0])}")
print(f"\nFirst element (query string) preview (first 200 chars):")
print(results[0][0][:200] if results[0][0] else "NONE")
print(f"\nSecond element (candidates) type: {type(results[0][1])}")
print(f"Second element (candidates) length: {len(results[0][1]) if isinstance(results[0][1], list) else 'N/A'}")
if isinstance(results[0][1], list) and len(results[0][1]) > 0:
    print(f"First candidate: {results[0][1][0][:100] if isinstance(results[0][1][0], str) else results[0][1][0]}")

def extract_entity_from_string(text_string):
    """
    Extract entity URI or BNode ID from a text string.
    Expected format: 'The entity <entity_identifier> has context ...'
    """
    match = re.search(r'The entity ([^\s]+)', text_string)
    if match:
        return match.group(1).strip()
    return None

def populate_target_column_in_dframe_origin(dframe_origin_df, similarity_results, target_column_name="aligned"):
    """
    Populate a specified column in dframe_origin_df based on similarity_results.
    dframe_origin_df: DataFrame containing NCIT entities.
    similarity_results: list of [DOID_query_string, [NCIT_candidate_strings]]
    target_column_name: The name of the column in dframe_origin_df to populate.
    """
    # Ensure the target column exists and is initialized with empty lists
    if target_column_name not in dframe_origin_df.columns:
        dframe_origin_df[target_column_name] = [[] for _ in range(len(dframe_origin_df))]
    else:
        # Re-initialize it to ensure it's empty for new population
        dframe_origin_df[target_column_name] = [[] for _ in range(len(dframe_origin_df))]

    # Create a temporary dictionary to store NCIT -> [DOID] alignments
    ncit_to_doid_alignments = defaultdict(list)

    processed_queries = 0
    skipped_queries = 0
    total_alignments_added = 0

    # Debugging print statements
    # print("--- Debug: Populating ncit_to_doid_alignments (first 3 candidates loop) ---")
    # candidate_loop_count = 0

    for item in similarity_results:
        if len(item) == 2:
            query_string = str(item[0])
            candidates_list = item[1] if isinstance(item[1], list) else []

            doid_entity = extract_entity_from_string(query_string)

            if doid_entity:
                processed_queries += 1
                for ncit_candidate_string in candidates_list:
                    ncit_entity = extract_entity_from_string(ncit_candidate_string)
                    if ncit_entity:
                        # if candidate_loop_count < 3:
                        #     print(f"  - Adding alignment: NCIT='{ncit_entity}' <= DOID='{doid_entity}'")
                        #     candidate_loop_count += 1
                        if doid_entity not in ncit_to_doid_alignments[ncit_entity]:
                            ncit_to_doid_alignments[ncit_entity].append(doid_entity)
                            total_alignments_added += 1
                    # else:
                        # print(f"  - Failed to extract NCIT entity from candidate string: {ncit_candidate_string[:100]}")
            # else:
                # print(f"  - Failed to extract DOID entity from query string: {query_string[:100]}")
                skipped_queries += 1
        # else:
            # print(f"  - Skipped item in similarity_results due to incorrect format: {item}")
            skipped_queries += 1

    # print(f"--- Debug: ncit_to_doid_alignments built. Total unique NCIT entities as keys: {len(ncit_to_doid_alignments)} ---")
    # if len(ncit_to_doid_alignments) > 0:
    #     first_key = next(iter(ncit_to_doid_alignments))
    #     print(f"First key in ncit_to_doid_alignments: '{first_key}' (type: {type(first_key)})")
    #     print(f"Sample value for first key: {ncit_to_doid_alignments[first_key][0]}")

    # print("\n--- Debug: Updating dframe_origin_df (first 3 lookups) ---")
    # df_update_count = 0
    for index, row in dframe_origin_df.iterrows():
        ncit_entity_uri_from_df = row['entity']
        # if df_update_count < 3:
        #     print(f"  - Looking up entity from dFrameOrigin: '{ncit_entity_uri_from_df}'")
        #     print(f"  - Is it in ncit_to_doid_alignments keys? {ncit_entity_uri_from_df in ncit_to_doid_alignments}")
        #     df_update_count += 1
        if ncit_entity_uri_from_df in ncit_to_doid_alignments:
            dframe_origin_df.at[index, target_column_name] = ncit_to_doid_alignments[ncit_entity_uri_from_df]

    # print("--- Debug: dframe_origin_df update loop finished ---")

    print(f"Processed {processed_queries} DOID queries, skipped {skipped_queries} queries due to parsing issues.")
    print(f"Total {total_alignments_added} unique DOID entities aligned as candidates to NCIT entities for column '{target_column_name}'.")

    return dframe_origin_df

# The function will be called in a subsequent cell to apply the changes.
```

```python
# First, ensure Ollama is installed
import subprocess
import sys

try:
    import ollama
except ImportError:
    print("Installing ollama package...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "ollama", "-q"])
    import ollama

def create_alignment_prompt(entity_uri, context, ncit_candidates, doid_candidates):
    """
    Create a prompt for Ollama to select the best aligned entity

    Args:
        entity_uri: The NCIT entity URI
        context: The context information about the entity
        ncit_candidates: List of NCIT candidate entities
        doid_candidates: List of DOID candidate entities

    Returns:
        prompt string for Ollama model
    """
    ncit_str = "\n".join([f"  {i+1}. {c}" for i, c in enumerate(ncit_candidates[:10])])
    doid_str = "\n".join([f"  {i+1}. {c}" for i, c in enumerate(doid_candidates[:10])])

    prompt = f"""You are an ontology alignment expert. Given an entity from NCIT (National Cancer Institute Thesaurus) and a list of candidate entities from DOID (Disease Ontology), select the BEST alignment candidate.

NCIT Entity: {entity_uri}
Context: {context[:5000]}

Top NCIT Candidates (for context):
{ncit_str if ncit_str else "  (none)"}

Top DOID Candidates (select from these):
{doid_str if doid_str else "  (none)"}

Based on the entity description and context, which DOID candidate is the BEST alignment?
Respond ONLY with the DOID entity URI (e.g., http://purl.obolibrary.org/obo/DOID_XXXX) or 'NO_MATCH' if none are suitable.

Answer:"""

    return prompt

def process_with_ollama(entity_uri, context, ncit_candidates, doid_candidates, model='qwen3', timeout=300):
    """
    Process entity alignment using Ollama model (default: qwen3)

    Args:
        entity_uri: The NCIT entity URI
        context: The context information
        ncit_candidates: NCIT candidates list
        doid_candidates: DOID candidates list
        model: Ollama model name
        timeout: Timeout in seconds

    Returns:
        Selected DOID entity URI or 'NO_MATCH'
    """
    try:
        prompt = create_alignment_prompt(entity_uri, context, ncit_candidates, doid_candidates)

        # Call Ollama with the prompt
        response = ollama.generate(
            model=model,
            prompt=prompt,
            stream=False,
            options={'num_predict': 5000, 'temperature': 0.2}  # Low temperature for consistency
        )

        # Extract the response
        answer = response['response'].strip()

        # Validate the answer is a URI or NO_MATCH
        if 'DOID' in answer or 'NO_MATCH' in answer:
            # Extract just the URI if there's extra text
            lines = answer.split('\n')
            for line in lines:
                if 'DOID' in line or 'NO_MATCH' in line:
                    answer = line.strip()
                    break
            return answer
        else:
            return 'NO_MATCH'

    except Exception as e:
        print(f"Error processing {entity_uri}: {str(e)}")
        return 'ERROR'

print("Ollama functions defined successfully")
print(f"Testing prompt creation with first entity...")
if len(dFrameOrigin) > 0:
    # Find the first row in dFrameOrigin that has populated 'Candidates'
    first_aligned_row_index = dFrameOrigin['Candidates'].apply(lambda x: len(x) > 0).first_valid_index()
    if first_aligned_row_index is not None:
        test_row = dFrameOrigin.loc[first_aligned_row_index]
        test_prompt = create_alignment_prompt(
            test_row['entity'],
            test_row['contextLevl0'][:2000], # Use contextLevl0 for NCIT entity context
            [], # No specific NCIT candidates needed for this test scenario, pass empty list
            test_row['Candidates'][:10] # Pass the list of DOID candidates directly
        )
        print("Sample prompt created (first 500 chars):")
        print(test_prompt[:5000])
    else:
        print("No entities in dFrameOrigin have populated 'Candidates' list to test the prompt.")
```

```python
def calculate_metrics(predicted_df, gold_standard_df):
    """
    Calculate Precision, Recall, and F1-Score

    Args:
        predicted_df: DataFrame with 'entity' and 'aligned' columns
        gold_standard_df: DataFrame with 'SrcEntity' and 'TgtEntity' columns

    Returns:
        Dictionary with metrics and detailed results
    """

    # Create mapping from gold standard
    gold_standard = {}
    for idx, row in gold_standard_df.iterrows():
        src = row['SrcEntity']
        tgt = row['TgtEntity']
        if src not in gold_standard:
            gold_standard[src] = []
        gold_standard[src].append(tgt)

    # Calculate metrics
    true_positives = 0
    false_positives = 0
    false_negatives = 0

    detailed_results = []

    # Check predictions
    for idx, row in predicted_df.iterrows():
        entity = row['entity']
        predicted = row['aligned']

        # Skip error cases
        if predicted in ['ERROR', 'error', None]:
            predicted = 'NO_MATCH'

        if entity in gold_standard:
            gold_targets = gold_standard[entity]

            # Check if prediction is in gold standard
            if predicted in gold_targets:
                true_positives += 1
                status = 'TP'
            elif predicted != 'NO_MATCH':
                false_positives += 1
                status = 'FP'
            else:
                false_negatives += 1
                status = 'FN'
        else:
            # Entity not in gold standard
            if predicted != 'NO_MATCH':
                false_positives += 1
                status = 'FP'
            else:
                status = 'TN'  # True negative

        detailed_results.append({
            'entity': entity,
            'predicted': predicted,
            'gold_standard': gold_standard.get(entity, []),
            'status': status
        })

    # Add false negatives for entities in gold standard but not predicted correctly (if any missing from df)
    # The previous loop covers keys in predicted_df. If predicted_df covers all entities, this block handles leftovers?
    # In this dataset structure, dFrameOrigin contains all source entities.

    # Calculate metrics
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return {
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'true_positives': true_positives,
        'false_positives': false_positives,
        'false_negatives': false_negatives,
        'total_predictions': len(predicted_df),
        'detailed_results': detailed_results
    }
```
